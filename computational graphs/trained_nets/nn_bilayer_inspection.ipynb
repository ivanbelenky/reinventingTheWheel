```python
import matplotlib.pyplot as plt
import pickle
import cv2
```

```python
import sys

fp = sys.path[0]+'/nn_bilayer_mnist.modlib'

with open(fp,"rb") as file:
    weight_list = pickle.load(file) 
```

```python
W1,W2 = weight_list
```

```python
W1 = W1[:,:-1]
W2 = W2[:,:-1]
```

```python
images = [neuron.reshape((28,28)) for neuron in W1]
```

```python
import numpy as np
images_renormalized = [np.round((255*(im-np.min(im))/np.max(im-np.min(im)))) for im in images]

```

```python
dotted = W2.dot(W1)
images_dotted = [neuron.reshape(28,28) for neuron in dotted]
images_dotted_renormalized = [np.round(255*(im-np.min(im)/np.max(im-np.min(im)))) for im in images_dotted]
```

```python
fig,axes = plt.subplots(figsize=(20,50),nrows=10,ncols=5)
axes = axes.reshape(-1)

for im,ax in zip(images_renormalized,axes):
    imshow = ax.imshow(im.astype('uint8'))
```

```python
fig,axes = plt.subplots(figsize=(20,10),nrows=2,ncols=5)
axes = axes.reshape(-1)

for im,ax in zip(images_dotted_renormalized,axes):
    ax.imshow(im.astype('uint8'))
```
